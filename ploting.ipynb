```python
import pandas as pd
data = pd.read_csv("DataCyclesAndEpisodeTime.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
summary = pd.DataFrame(data.groupby('Work').mean())
summary1 = pd.DataFrame(data.groupby('Work').sum())
summary2 = pd.DataFrame(data.groupby('Work').median())
data["episode duration in minutes"] = data["episode duration in seconds"].apply(lambda x: x/60 )
```

```python
l=sns.lmplot(x="episode duration in minutes", y="number of edit-run cycles", hue="Work",markers=["o", "x"],truncate=False, data=data, legend_out=False)
plt.legend(bbox_to_anchor=(0.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()
plt.savefig("episodeTimeVsCycle.pdf")
```

```python
sns.relplot(
    data=data, x="episode duration in seconds", y="number of edit-run cycles",
    col="Work", hue="Work", style="Work",
    kind="scatter"
)
plt.ylim(0,None)
plt.xlim(0,None)
```

```python
# this is correlation for the number of cycles and progamming time
debugging = data.loc[(data.Work == "Debugging")]
rDebugging = debugging["episode duration in minutes"].corr(debugging["number of edit-run cycles"])
rDebugging
```

```python
# this is correlation for the number of cycles and progamming time
programming = data.loc[(data.Work == "Programming")]
rProgramming = programming["episode duration in minutes"].corr(programming["number of edit-run cycles"])
rProgramming
```
